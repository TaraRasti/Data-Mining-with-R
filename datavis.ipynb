```r
plotdata <- df1 %>%
  count(frame_type) %>%
  arrange(desc(frame_type)) %>%
  mutate(prop = round(n*100/sum(n), 1),
         lab.ypos = cumsum(prop) - 0.5*prop) 

plotdata$label <- paste0(plotdata$frame_type, "\n",
                         round(plotdata$prop),"%")
ggplot(plotdata, aes(x = "",
                     y = prop,
                     fill = frame_type)) + 
  geom_bar(width = 1,
           stat = "identity",
           color = "black") + 
  geom_text(aes(y = lab.ypos, label = label),
            color = "black") + 
  coord_polar("y",
              start = 0,
              direction = -1) + 
  theme_void() +
  theme(legend.position = "FALSE") + labs(title = "Frame Type")
 
```
