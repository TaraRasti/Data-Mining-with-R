```r
library(forecast)
library(jtools)
library(huxtable)
```

```r
set.seed(1)
train.index=sample(c(1:dim(data)[1]), dim(data)[1]*0.6)
train.df=data[train.index,]
valid.df=data[-train.index,]
```

```r

logit.reg=glm(Dahak~.,data=train.df,family = 'binomial')
summary(logit.reg)
summ(logit.reg)
```

```r
logit.reg.step=step(logit.reg, direction = "backward")
summary(logit.reg.step)
summ(logit.reg.step)
```

```r
library(caret)
valid.pred = predict(logit.reg.step, valid.df)
train.pred = predict(logit.reg.step, train.df)
train.pred = ifelse(train.pred > 0.5, 1, 0)
valid.pred = ifelse(valid.pred > 0.5, 1, 0)
confusionMatrix(as.factor(train.pred), as.factor(train.df$Dahak))
confusionMatrix(as.factor(valid.pred), as.factor(valid.df$Dahak))
```
